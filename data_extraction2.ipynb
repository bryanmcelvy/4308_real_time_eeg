```python
from mne.io import read_raw_edf
import numpy as np
import os
import pandas as pd
```

```python
main_path = '/Users/bryanmcelvy/Documents/physionet.org/files/chbmit/1.0.0'

patient_data = pd.DataFrame(columns="start,end".split(','))
patient_data.rename_axis("filename", axis="index")
patient_data.head()
```

```python
with open(f"{main_path}/RECORDS") as file:
    records = file.read()
    
records = records.split()
records[:5]
```

```python
patient_data = pd.DataFrame()

for file in records:
    filename = file
    full_file_path = f"{main_path}/{filename}"; print(full_file_path)
    temp_file = read_raw_edf(input_fname = full_file_path, preload=False, verbose='ERROR')
    temp_df = pd.DataFrame(data=np.transpose(temp_file.get_data()), 
                        columns=temp_file.ch_names, index=[filename for _ in range(len(temp_file))]
                        )
    temp_df.insert(loc=0, column='Time', value = np.arange(len(temp_file)) / temp_file.info["sfreq"])
    temp_df.rename_axis("filename", axis="index", inplace=True)
    patient_data = pd.concat([patient_data, temp_df])
```
